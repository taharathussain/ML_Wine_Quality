### Plotting histogram for each feature for better understanding

```python
attribute_only = wine_data.drop(['quality'], axis=1)
plt.rcParams["figure.figsize"] = (16,10)              
attribute_only.hist()
plt.show()
```

### Checking correlation between columns

```python
plt.rcParams["figure.figsize"] = (15,10)
sns.heatmap(wine_data.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (8,2)
```

```python
# Plotting each feature to check outliers
Column_names =['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
for col in Column_names:
    sns.boxplot(x=col,data=wine_data)
    plt.show()
```

```python
# Writing a function to remove outliers

def outlier_removed (c):
    
    a = wine_data[c].mean() + 3*wine_data[c].std()
    b = wine_data[c].mean() - 3*wine_data[c].std()
    wine_data[c] = np.where(
    wine_data[c]>a,
    wine_data[c].mean(),
    np.where(
        wine_data[c]< b,
        wine_data[c].mean(),
        wine_data[c]
    )
)
```

```python
#Removing outliers
for col in Column_names:
    outlier_removed(col)
```

```python
#Checking boxplot after removing outlier
sns.boxplot(x='residual sugar',data=wine_data)
```

### Looking at unique value for Dependent variables 

```python
wine_data['quality'].value_counts()
```

### Mapping Quality to three classes (0 = Low quality, 1 = Medium quality and 2 = Good quality)

```python
bins_size = [0, 5.5, 7.5, 10]
label = [0, 1, 2]
wine_data['quality'] = pd.cut(wine_data['quality'], bins=bins_size, labels=label)
```

```python
wine_data['quality'].value_counts()
```

### Checking relationship of  features with  Quality

```python
plt.rcParams["figure.figsize"] = (8,2)
for col in Column_names:
    sns.barplot(x=col,y = 'quality', data=wine_data)
    plt.show()
```

# Data Preparation for model building

```python
# Defining Features X
X = attribute_only
```

```python
# Defining target y
y = wine_data[['quality']]
```

```python
# Data Standardization
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
# Splitting data into train and test 

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

### Model 1 - DecisionTree Classifier

```python
# Defining Model
DT = DecisionTreeClassifier(random_state=5)
```

```python
# Fitting Model
DT.fit(X_train, y_train)
```

```python
# Prediction for test data
Predicted_DT = DT.predict(X_test)
```

```python
# Checking performance of Model
accuracy_score(y_test,Predicted_DT)
```

```python
print(metrics.classification_report(y_test, Predicted_DT))
```

```python
# Prediction of Train data
Predicted_DT2 = DT.predict(X_train)
```

```python
# Checking Accuracy for Train dat a
accuracy_score(y_train,Predicted_DT2)
```

### Model 2 - Logistic Regression

```python
# Defining Model
LR = LogisticRegression(multi_class='multinomial',solver ='newton-cg')
```

```python
# Fitting Model
LR.fit(X_train, y_train)
```

```python
# Prediction for test data
Predicted_LR = LR.predict(X_test)
```

```python
# Checking performance of Model
accuracy_score(y_test,Predicted_LR)
```

```python
print(metrics.classification_report(y_test, Predicted_LR))
```

### Model 3 - Random Forest Classifier

```python
# Defining Model
RF = RandomForestClassifier()
```

```python
# Fitting Model
RF.fit(X_train, y_train)
```

```python
# Prediction for test data
Predicted_RF = RF.predict(X_test)
```

```python
# Checking performance of Model
accuracy_score(y_test,Predicted_RF)
```

```python
print(metrics.classification_report(y_test, Predicted_RF))
```
